```python
sigmoid_vec = np.arange(-10, 10, 0.5)

def sigmoid(lst):
  sigmoid_ans = []
  for i in lst:
    x = 1 / (1 + np.exp(-i))
    sigmoid_ans.append(x)
  return sigmoid_ans

sigmoid_answer = sigmoid(sigmoid_vec)
reshaped = np.reshape(sigmoid_answer, (5, 8))
print(reshaped)
```

```python
import os
from PIL import Image

def is_image_file(file_path):
    supported_extensions = {'.png', '.jpeg', '.jpg', '.tiff'}
    _, extension = os.path.splitext(file_path)
    return extension.lower() in supported_extensions

def find_image_files(folder_path):
    image_files = []

    for root, _, files in os.walk(folder_path):
        for file in files:
            file_path = os.path.join(root, file)
            if is_image_file(file_path):
                image_files.append(file_path)

    return image_files


folder_path = 'path'
image_files = find_image_files(folder_path)


if image_files:
    print("Image Files:")
    for image_file in image_files:
        print(image_file)
else:
    print("No image files found.")
```

```python
class Rectangle:
  def __init__(self, length, breadth):
    self.length = length
    self.breadth = breadth

  def area (self):
    return self.length * self.breadth

  def perimeter (self):
    return 2*(self.length + self.breadth)

length = int(input('length = '))
breadth = int(input('breadth = '))
rect = Rectangle(length, breadth)
print(rect.area())
print(rect.perimeter())
```

```python
import requests
from bs4 import BeautifulSoup
import csv

def extract_paper_details(url):
    # Send a GET request to the specified URL
    response = requests.get(url)

    # Check if the request was successful (status code 200)
    if response.status_code == 200:
        # Parse the HTML content using BeautifulSoup
        soup = BeautifulSoup(response.text, 'html.parser')

        # Find all elements with the class 'ptitle' representing paper titles
        papers = soup.find_all('dt', {'class': 'ptitle'})

        # Extract details for each paper
        data = []
        for paper in papers:
            # Extract title, authors, and PDF link
            title = paper.find('a').text.strip()
            authors = paper.find_next('dd').text.replace('\n', '').strip()
            pdf_link = paper.find('a')['href']

            # Append the extracted data to the list
            data.append([authors, title, pdf_link])

        return data
    else:
        # Print an error message if the request fails
        print(f"Failed to fetch the page. Status Code: {response.status_code}")
        return None

def write_to_csv(data, csv_filename='papers_data.csv'):
    # Write the extracted data to a CSV file
    with open(csv_filename, mode='w', newline='', encoding='utf-8') as file:
        writer = csv.writer(file)

        # Write the header row
        writer.writerow(['Authors', 'Title', 'PDF'])

        # Write each row of data to the CSV file
        for row in data:
            writer.writerow(row)

if __name__ == "__main__":
    # URL of the webpage containing paper details
    target_url = 'https://openaccess.thecvf.com/ICCV2023?day=all'

    # Extract paper details from the webpage
    paper_data = extract_paper_details(target_url)

    # Check if data is successfully extracted
    if paper_data:
        print("Extracted data:")
        for row in paper_data:
            print(row)

        # Write the extracted data to a CSV file
        write_to_csv(paper_data, 'papers_data.csv')
        print("Data written to CSV file: papers_data.csv")
```
